```python
def gender_features(word):
    return {'last_letter': word[-1]}
```

```python
from nltk.corpus import names
```

```python
labeled_names = ([(name,'male') for name in names.words('male.txt')] +
                 [(name,'female') for name in names.words('female.txt')])
len(labeled_names)
```

```python
import random
random.shuffle(labeled_names)
```

```python
featureset = [(gender_features(n), gender) for (n,gender) in labeled_names]
```

```python
train_set, test_set = featureset[400:],featureset[:400]
```

```python
import nltk
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
classifier.classify(gender_features('David'))
```

```python
classifier.classify(gender_features('Michelle'))
```

```python
classifier.classify(gender_features('Viji'))
```

```python
classifier.classify(gender_features('Santhosh'))
```

```python
print(nltk.classify.accuracy(classifier, test_set))
```
